```python
url = 'https://data.gov.il/api/3/action/datastore_search?resource_id=053cea08-09bc-40ec-8f7a-156f0677aff3'
import requests
import pandas as pd
pd.set_option("display.max_columns", None)
response = requests.get(url)
data = response.json()
df = pd.DataFrame(data)

data_df = pd.DataFrame(data['result']['records'])

#יציג את 5 השורות הראשונות של הנתונים

data_df.head()
```

```python
from ipywidgets import Dropdown, VBox
from IPython.display import display

# Initialize the dropdown widgets
Tozeret = Dropdown(
    options=data_df['tozeret_nm'].dropna().unique().tolist(),
    description='Tozeret:',
)

Kinui = Dropdown(
    options=[],  # Empty initially
    description='Kinuy Mishari:',
)

# Flag to control if Kinui's observer should execute
kinui_ready = False

# Function to update Kinui options based on Tozeret selection
def on_selection_change_tozeret(change):
    global kinui_ready
    selected_tozeret = change['new']
    # Filter the Kinui options based on selected Tozeret
    filtered_kinui = data_df[data_df['tozeret_nm'] == selected_tozeret]['kinuy_mishari'].dropna().unique().tolist()
    Kinui.options = filtered_kinui
    Kinui.value = None  # Reset Kinui's value
    kinui_ready = True  # Allow Kinui's observer to execute after this

# Function to handle Kinui selection and print filtered info
def on_selection_change_Kinui(change):
    global kinui_ready
    if not kinui_ready or change['new'] is None:  # Ensure the observer runs only when ready
        return
    selected_kinui = change['new']
    selected_tozeret = Tozeret.value
    if selected_tozeret and selected_kinui:  # Ensure valid selections
        filtered_df = data_df[(data_df['tozeret_nm'] == selected_tozeret) &
                              (data_df['kinuy_mishari'] == selected_kinui)]
        filtered_count = len(filtered_df)
        unique_ramat_gimur = filtered_df['ramat_gimur'].dropna().unique().tolist()
        print(f"Total records: {filtered_count}")
        print(f"Unique 'ramat_gimur': {unique_ramat_gimur}")

# Attach observers
Tozeret.observe(on_selection_change_tozeret, names='value')
Kinui.observe(on_selection_change_Kinui, names='value')

# Display the UI
ui = VBox([Tozeret, Kinui])
display(ui)
```
